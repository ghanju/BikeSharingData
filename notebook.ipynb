## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana tren peminjaman sepeda pada rentang tahun 2011 sampai 2012?


```python
main = hour.copy()
mappingYr = {0: '2011', 1: '2012'}
mappingMo = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
                7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
mappingSeas = {1:'Spring', 2: 'Summer', 3:'Fall', 4:'Winter'}
mappingWeath = {1:'Clear', 2: 'Cloudy', 3:'Light Snow/Rain', 4:'Heavy Rain/Ice Pallets'}


main['year'] = main['year'].map(mappingYr)
main['month'] = main['month'].map(mappingMo)
main['weekday'] = main['date'].dt.day_name()
main['season'] = main['season'].map(mappingSeas)
main['weathersit'] = main['weathersit'].map(mappingWeath)
main = main[['date', 'year', 'month', 'hour', 'weekday', 'season', 'weathersit','total_user', 'registered_user', 'casual_user']]

print(main)
```

Output diatas, kita menjadikan value setiap kolom lebih jelas untuk kepentingan analisi lebih lanjut dan lebih jelas dalam pemvisualisasian

```python
new_monthly_data = main.groupby(['year', 'month'])[['total_user', 'registered_user', 'casual_user']].sum().reset_index()

new_monthly_data = new_monthly_data[['year', 'month', 'total_user', 'registered_user', 'casual_user']]

print(new_monthly_data)
```

```python
new_monthly_data['date'] = new_monthly_data['year'] + '-' + new_monthly_data['month']
new_monthly_data['date'] = pd.to_datetime(new_monthly_data['date'], format='%Y-%B')
new_monthly_data = new_monthly_data.sort_values('date')

plt.figure(figsize=(14, 8))
plt.plot(new_monthly_data['date'], new_monthly_data['total_user'], marker='o', label='Total Rentals (total_user)', color='blue')
plt.title('Peminjaman Sepeda Bulanan (2 tahun)')
plt.xlabel('Date')
plt.ylabel('Number of Rentals')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

Tren peminjaman sepeda selama 2 tahun memiliki pola "Meningkat pada Q1 dan menurun pada Q4". Dengan melihat grafik diatas dapat dikatakan penyewaan akan meningkat drastis pada awal tahun dan terus meningkat sampai kepada tengah tahun. Lalu pada Q3 tahun tersebut penyewaan akan menurun. Namun, dalam dua tahun penyewaan memiliki peningkatan pada dari tahun ke tahun

### Pertanyaan 2:
Berapa perbandingan banyaknya peminjam dari peminjem biasa dan pelanggan?

```python
registered_total = main['registered_user'].sum()
casual_total = main['casual_user'].sum()
labels = ['Registered User', 'Casual User']
sizes = [registered_total, casual_total]
colors = ['Blue', 'Red']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.axis('equal')
plt.title('Persentase Perbandingan Registered Customer dan Casual Customer')
plt.tight_layout()
plt.show()
```

Penyewa sepeda lebih banyak berasal dari pelanggan (registered user) dengan persentase sebanyak 81,2% sementara dari peminjam biasa (casual user) hanya 18,8%

**Insight:**
- Tren yang terjadi dalam dua tahun penyewaan sepeda terjadi fluktuasi terhadap peminjaman pada setiap tahun. Kenaikan terjadi diawal tahun 2011 dan menurun diakhir tahun namun tidak seperti nilai awal, dan kembali naik 2012 melebihi penyewaan tertinggi pada tahun 2011 dan akhirnya menurun lagi. Penyewaan juga di dominasi oleh pelanggan yang terdaftar dan hanya sedikit dari pengguna biasa.

## Conclusion

- Tren peminjaman yang terjadi pada tahun 2011 sampai pada tahun 2012 mengalami peningkatan pada tahun berikut. Hal ini diikutin dengan tren naik pada awal tahun dan turun pada akhir tahun, karena pada Q1 itu melesat hingga Q2 lalu tibalah menurun pada Q3 hingga pada Q4.
- Perbandingan pelanggan dan peminjam biasa cukup signifikan. Penyewaan didominasi oleh pelanggan dengan rentang waktu sore hari, namun jika dilihat pada pengguna biasa penyewaan tinggi terjadi diantara jam 14-17. Hal ini bisa menjadi peluang kesempatan untuk menggapai orang-orang diluar pelanggan untuk bisa memperbesar use-time dari penyewaan sepeda.
